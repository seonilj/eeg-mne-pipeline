# Motor Imagery EEG Signal Enhancement
This notebook focuses on the secondary stage of the EEG pipeline: filtering and enhancing motor imagery signals (Mu/Beta rhythms) by suppressing line noise and physiological artifacts.

```python
# ==============================================================================
# Step 1 & 2: Environment Setup & Data Loading (Reused from Notebook 1)
# ==============================================================================
!pip install mne matplotlib
import mne

# Load the exact same Subject 1, Run 4 data
subject = 1
runs = [4]
raw_fnames = mne.datasets.eegbci.load_data(subject, runs)
raw = mne.io.read_raw_edf(raw_fnames[0], preload=True)

# Clean channel names to match standard repository format
mapping = {ch_name: ch_name.rstrip('.') for ch_name in raw.ch_names}
raw.rename_channels(mapping)
```

```python
# ==============================================================================
# Step 6: Power-Line Noise Removal via Notch Filtering (60Hz)
# ==============================================================================

print("Applying Notch Filter to remove 60Hz power-line noise (Hum noise)")

# The dataset is recorded in a 60Hz power-line environment. We suppress this frequency component.
raw_notch = raw.copy().notch_filter(freqs=60.0, verbose=False)
```

```python
# ==============================================================================
# Step 7: Band-Pass Filtering for Motor Imagery Isolation (1-40Hz)
# ==============================================================================

print("Executing Band-Pass Filtering (1.0Hz - 40.0Hz)")

# Isolate 8-30Hz (Mu/Beta rhythms) over the motor cortex while removing low/high-frequency artifacts
raw_filtered = raw_notch.copy().filter(l_freq=1.0, h_freq=40.0, verbose=False)

print("\nSignal enhancement pipeline built successfully!")
```
